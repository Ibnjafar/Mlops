## 2 . Loading VGG Model

```python
from keras.applications import vgg16
# pre trained weights : create my model vgg16
model = vgg16.VGG16(weights='imagenet')
```

```python
model.layers[0].input
```

## 3. Loading and Preprocessing image

```python
from keras.preprocessing import image
img = image.load_img('img/cat.jpg', target_size=(224,224))
```

```python
img
```

```python
img.size
```

```python
type(img)
```

```python
img_np = image.img_to_array(img)
```

```python
img_np.shape
```

```python
type(img_np)
```

```python
import numpy as np
```

```python
ae = np.expand_dims(img_np, axis=0)
```

```python
ae.shape
```

## 4. Prediction 

```python
from keras.applications.vgg16 import decode_predictions
```

```python
from keras.applications.vgg16 import preprocess_input
```

```python
finalimg = preprocess_input(ae)
```

```python
pred= model.predict(finalimg)
```

```python
label=decode_predictions(pred)
```

```python
label
```

```python
label = label[0][0]
# print the classification
print('%s (%.2f%%)' % (label[1], label[2]*100))
```
